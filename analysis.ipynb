```python
df=df.loc[:,~df.columns.str.contains('^Unnamed')]   #remove unwanted Unnamed colunms
df.columns
```

```python
budget_2022_23=df[['Particulars','Budget 2022-2023 Revenue','Budget 2022-2023 Capital','Budget 2022-2023 Total']]
budget_2022_23.head()
```

```python
budget_2022_23.loc[:, budget_2022_23.columns[1:]] = (
    budget_2022_23.loc[:, budget_2022_23.columns[1:]]
    .apply(pd.to_numeric, errors='coerce')
)
```

```python
import numpy as np

col = 'Budget 2022-2023 Total'

budget_2022_23[col] = (
    budget_2022_23[col]
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace('-', '', regex=False)
    .str.strip()
)

budget_2022_23[col] = pd.to_numeric(budget_2022_23[col], errors='coerce')
```

```python
top10 = (
    budget_2022_23[['Particulars', col]]
    .dropna()
    .sort_values(by=col, ascending=False)
    .head(10)
)

top10
```
